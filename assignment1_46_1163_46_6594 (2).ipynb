# **- Given Functions**

```python
#Function 1: Given an image, returns the image and it's squared version in array format
def toArrayAndSquare(im):
  img = [[0 for x in range(im.size[0])] for y in range(im.size[1])]
  sqr = [[0 for x in range(im.size[0])] for y in range(im.size[1])]
  for i in range (0,im.size[0]):
    for j in range (0,im.size[1]):
      img[j][i] = im.getpixel((i,j))
      sqr[j][i] = img[j][i] ** 2
  return img,sqr

#Function 2: Given an image, applies a certain threshold
def applyThreshold (im, th):
  res = [[0 for x in range(len(im[0]))] for y in range(len(im))]
  for i in range (0,len(im)):
    for j in range (0,len(im[0])):
      if(im[i][j]<th):
        res[i][j] = 0
      else:
        res[i][j] = 255
  return res
```

```python
def imgWithCooccurrence(comatrix1,comatrix2,image):

    borders = [[0 for x in range(image.size[0])] for y in range(image.size[1])]
    imA=np.array(image)
    for i in range (0,len(imA)-1):
      for j in range (0,len(imA[0])-1):
            if(comatrix1[imA[i][j]][imA[i+1][j]] != 0):
                borders[i][j] = 255
            if(comatrix2[imA[i][j]][imA[i][j+1]] != 0):
                borders[i][j] = 255


    return borders
```

```python
def integralArray(imgarray):
  img = np.zeros((480,720))

  for i in range (0,len(imgarray)):
    img[0][i] = imgarray[0][i]

  for i in range (1,len(imgarray)):
    for j in range (0,len(imgarray[0])):
      img[i][j] = imgarray[i][j] + img[i-1][j]

  for i in range (0, len(imgarray)):
    for j in range (1, len(imgarray[0])):
      img[i][j] = img[i][j] + img[i][j-1]

  return img
```

```python
def localSum(int,i1,j1,i2,j2):
  localsum = 0
  if(i1 == 0 and j1 == 0):
    localsum = int[i2][j2]
  if(i1 == 0 and j1 > 0):
    localsum = int[i2][j2]-int[i2][j1-1]
  if(i1 > 0 and j1 == 0):
    localsum = int[i2][j2]-int[i1-1][j2]
  if(i1 > 0 and j1 > 0):
    localsum = int[i2][j2]-int[i2][j1-1]-int[i1-1][j2]+int[i1-1][j1-1]

  return localsum
```

```python
def imgWithIntegral(integral1, integral2, x, y):
  vararray = np.zeros((480,720))
  locsum1 = 0
  locsum2 = 0
  variance = 0

  if(x > 480):
    x = 480
  if(y > 720):
    y = 720

  for i in range (1,len(integral1)-1):
    for j in range (1,len(integral1[1])-1):
      locsum1 = localSum(integral1,i-1,j-1,i+1,j+1)
      locsum2 = localSum(integral2,i-1,j-1,i+1,j+1)
      variance = (((1/(x*y))*locsum2) - (((1/(x*y))*locsum1)**2))
      if(variance < 0):
        vararray[i][j] = 0
      else:
        vararray[i][j] = variance

  return vararray
```

Comments:
Co-occurence reconstruction results in better output than Integral image reconstruction because lines are more defined in its output.
This can be attributed to the fact that co-occurence matrices contain details about the relationship between pixels while this is not true for the integral image
Complexity:
Reconstruction with Co-occurence matrix has a complexity of O(m*n) where m is rows and n is columns of image, calculating comatrices for the image takes 2 seconds of runtime O(m*n), while reconstruction takes less than 1 second.
Reconstruction with Integral images has same complexity but lower efficieny as it takes more time to compute, and results in a much lower quality reconstruction.
